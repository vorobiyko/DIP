### Імпортуємо необхдіні пакети та бібліотеки

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

# Завдання
## 1. Порівняти гістограми двох зображень.

1.1. Відкрити та завантажити два зображення:
 

```python
img_half = cv2.imread("IN\Lab3_half100.png", cv2.IMREAD_GRAYSCALE)
img_chess = cv2.imread("IN\Lab3_chess100.png", cv2.IMREAD_GRAYSCALE)


plt.figure(figsize=(7, 3))
plt.subplot(121), plt.imshow(img_half, cmap = 'gray'), plt.xticks([]), plt.yticks([]), plt.title('half')
plt.subplot(122), plt.imshow(img_chess, cmap = 'gray'), plt.xticks([]), plt.yticks([]), plt.title('chess')

plt.show();
```

1.2.Визначити гістограму для кожного зображення:

```python
hist_half = cv2.calcHist([img_half], [0], None, [256], [0, 256])
hist_chess = cv2.calcHist([img_chess], [0], None, [256], [0, 256])
```

1.3.Відобразити завантаженні зображення та відповідно визначені гістограми зображень за допомогою функції **plt.subplot**:

```python
plt.figure(figsize=(7, 3))
plt.subplot(121), plt.imshow(img_half, cmap = 'gray'), plt.xticks([]), plt.yticks([]), plt.title('half')
plt.subplot(122), plt.plot(hist_half), plt.xticks([]), plt.yticks([]), plt.title('half - gistogram')

plt.figure(figsize=(7, 3))
plt.subplot(121), plt.imshow(img_chess, cmap = 'gray'), plt.xticks([]), plt.yticks([]), plt.title('chess')
plt.subplot(122), plt.plot(hist_chess), plt.xticks([]), plt.yticks([]), plt.title('chess - gistogram')

plt.show()
```

1.4 Виконати просторову фільтрацію зображень використовуючи згладжувальний фільтра або розмивання Гауса з  розмірами - 3x3, 5x5, 9x9, 15x15.

```python
half_blur3 = cv2.GaussianBlur(img_half,(3,3),0)
half_blur5 = cv2.GaussianBlur(img_half,(5,5),0)
half_blur9 = cv2.GaussianBlur(img_half,(9,9),0)
half_blur15 = cv2.GaussianBlur(img_half,(15,15),0)

chess_blur3 = cv2.GaussianBlur(img_chess,(3,3),0)
chess_blur5 = cv2.GaussianBlur(img_chess,(5,5),0)
chess_blur9 = cv2.GaussianBlur(img_chess,(9,9),0)
chess_blur15 = cv2.GaussianBlur(img_chess,(15,15),0)
```

1.5 Визначити гістограми для всіх отриманих зображень.

```python
half_hist = cv2.calcHist([img_half], [0], None, [256], [0, 256])
half_blur3_hist = cv2.calcHist([half_blur3], [0], None, [256], [0, 256])
half_blur5_hist = cv2.calcHist([half_blur5], [0], None, [256], [0, 256])
half_blur9_hist = cv2.calcHist([half_blur9], [0], None, [256], [0, 256])
half_blur15_hist = cv2.calcHist([half_blur15], [0], None, [256], [0, 256])

chess_hist = cv2.calcHist([img_chess], [0], None, [256], [0, 256])
chess_blur3_hist = cv2.calcHist([chess_blur3], [0], None, [256], [0, 256])
chess_blur5_hist = cv2.calcHist([chess_blur5], [0], None, [256], [0, 256])
chess_blur9_hist = cv2.calcHist([chess_blur9], [0], None, [256], [0, 256])
chess_blur15_hist = cv2.calcHist([chess_blur15], [0], None, [256], [0, 256])
```

1.6.Відобразити отримані після фільтрації зображення та відповідно визначені гістограми за допомогою функції plt.subplot:

```python
plt.figure(figsize=(10,20))
plt.subplot(521), plt.imshow(img_half, cmap='gray'), plt.title('Original'),plt.xticks([]), plt.yticks([])
plt.subplot(523), plt.imshow(half_blur3, cmap='gray'),  plt.title('Gaussian Blur 3x3'),   plt.xticks([]), plt.yticks([])
plt.subplot(525), plt.imshow(half_blur5, cmap='gray'),  plt.title('Gaussian Blur 5x5'),   plt.xticks([]), plt.yticks([])
plt.subplot(527), plt.imshow(half_blur9, cmap='gray'),  plt.title('Gaussian Blur 9x9'),   plt.xticks([]), plt.yticks([])
plt.subplot(529), plt.imshow(half_blur15, cmap='gray'), plt.title('Gaussian Blur 15x15'), plt.xticks([]), plt.yticks([])
plt.subplot(522), plt.plot(half_hist),           plt.xlim([0, 255]), plt.grid(), plt.title('Original')
plt.subplot(524), plt.plot(half_blur3_hist),     plt.xlim([0, 255]), plt.grid(), plt.title('Blur 3x3')
plt.subplot(526), plt.plot(half_blur5_hist),     plt.xlim([0, 255]), plt.grid(), plt.title('Blur 5x5')
plt.subplot(528), plt.plot(half_blur9_hist),     plt.xlim([0, 255]), plt.grid(), plt.title('Blur 9x9')
plt.subplot(5,2,10), plt.plot(half_blur15_hist), plt.xlim([0, 255]), plt.grid(), plt.title('Blur 15x15')

plt.show()
```

```python
plt.figure(figsize=(10,20))
plt.subplot(521), plt.imshow(img_chess, cmap='gray'), plt.title('Original'),plt.xticks([]), plt.yticks([])
plt.subplot(523), plt.imshow(chess_blur3, cmap='gray'),  plt.title('Gaussian Blur 3x3'),   plt.xticks([]), plt.yticks([])
plt.subplot(525), plt.imshow(chess_blur5, cmap='gray'),  plt.title('Gaussian Blur 5x5'),   plt.xticks([]), plt.yticks([])
plt.subplot(527), plt.imshow(chess_blur9, cmap='gray'),  plt.title('Gaussian Blur 9x9'),   plt.xticks([]), plt.yticks([])
plt.subplot(529), plt.imshow(chess_blur15, cmap='gray'), plt.title('Gaussian Blur 15x15'), plt.xticks([]), plt.yticks([])
plt.subplot(522), plt.plot(chess_hist),           plt.xlim([0, 255]), plt.grid(), plt.title('Original')
plt.subplot(524), plt.plot(chess_blur3_hist),     plt.xlim([0, 255]), plt.grid(), plt.title('Blur 3x3')
plt.subplot(526), plt.plot(chess_blur5_hist),     plt.xlim([0, 255]), plt.grid(), plt.title('Blur 5x5')
plt.subplot(528), plt.plot(chess_blur9_hist),     plt.xlim([0, 255]), plt.grid(), plt.title('Blur 9x9')
plt.subplot(5,2,10), plt.plot(chess_blur15_hist), plt.xlim([0, 255]), plt.grid(), plt.title('Blur 15x15')

plt.show()
```
