Завантажуємо бібліотеки.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# 1. Зчитати та завантажити тестове чорно-біле та кольорове зображення.

### Зчитаємо тестове кольорове зображення та конвертуєму його в чорно-біле та кольрові зображення

```python
img_bgr = cv2.imread(r"IN\its_me.jpg")
img_gray = cv2.imread(r"IN\its_me.jpg", cv2.IMREAD_GRAYSCALE) 

cv2.imwrite("OUT\image_bgr.png", img_bgr)
cv2.imwrite("OUT\image_gray.png", img_gray)
```

## 2. Відобразити чорно-біле та кольорове зображення

Відображення зображень за допомогою matplotlib

```python
plt.figure(figsize=(8, 3))
plt.subplot(121), plt.imshow(img_gray, cmap='gray'), plt.xticks([]), plt.yticks([]), plt.title("GRAY")
plt.subplot(122), plt.imshow(img_bgr), plt.xticks([]), plt.yticks([]), plt.title("BGR")

plt.show();
```

## 3. Перетворити зображення з колірної моделі BGR в колірні моделі RGB і HSV та відобразити.

```python
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
img_hsv = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2HSV)

plt.figure(figsize=(8, 3))
plt.subplot(121), plt.imshow(img_rgb), plt.xticks([]), plt.yticks([]), plt.title("RGB")
plt.subplot(122), plt.imshow(img_hsv), plt.xticks([]), plt.yticks([]), plt.title("HSV")
plt.show();

cv2.imwrite("OUT\image_rgb.png", img_rgb)
cv2.imwrite("OUT\image_hsv.png", img_hsv)
```

## 4. Окремо відобразити колірні канали для зображень із колірними моделями RGB та HSV.

За допомогою індексу

```python
R = np.zeros(img_rgb.shape, dtype='uint8')
G = np.zeros(img_rgb.shape, dtype='uint8')
B = np.zeros(img_rgb.shape, dtype='uint8')

H = np.zeros(img_hsv.shape, dtype='uint8')
S = np.zeros(img_hsv.shape, dtype='uint8')
V = np.zeros(img_hsv.shape, dtype='uint8')

R[:, :, 0] = img_rgb[:, :, 0]
G[:, :, 1] = img_rgb[:, :, 1]
B[:, :, 2] = img_rgb[:, :, 2]

H[:, :, 0] = img_hsv[:, :, 0]
S[:, :, 1] = img_hsv[:, :, 1]
V[:, :, 2] = img_hsv[:, :, 2]

plt.figure(figsize=(10, 6))
plt.subplot(231), plt.imshow(R), plt.xticks([]), plt.yticks([]), plt.title("RED")
plt.subplot(232), plt.imshow(G), plt.xticks([]), plt.yticks([]), plt.title("GREEN")
plt.subplot(233), plt.imshow(B), plt.xticks([]), plt.yticks([]), plt.title("BLUE")

plt.subplot(234), plt.imshow(H), plt.xticks([]), plt.yticks([]), plt.title("NUE")
plt.subplot(235), plt.imshow(S), plt.xticks([]), plt.yticks([]), plt.title("SATURATION")
plt.subplot(236), plt.imshow(V), plt.xticks([]), plt.yticks([]), plt.title("VALUE")

plt.show()
```

За допомогою функції split

```python
R = np.zeros(img_rgb.shape, dtype='uint8')
G = np.zeros(img_rgb.shape, dtype='uint8')
B = np.zeros(img_rgb.shape, dtype='uint8')

H = np.zeros(img_hsv.shape, dtype='uint8')
S = np.zeros(img_hsv.shape, dtype='uint8')
V = np.zeros(img_hsv.shape, dtype='uint8')

b, g, r = cv2.split(img_rgb)
h, s, v = cv2.split(img_hsv)

R[:, :, 0] = r
G[:, :, 1] = g
B[:, :, 2] = b

H[:, :, 0] = h
S[:, :, 1] = s
V[:, :, 2] = v

plt.figure(figsize=(10, 6))
plt.subplot(231), plt.imshow(R), plt.xticks([]), plt.yticks([]), plt.title("RED")
plt.subplot(232), plt.imshow(G), plt.xticks([]), plt.yticks([]), plt.title("GREEN")
plt.subplot(233), plt.imshow(B), plt.xticks([]), plt.yticks([]), plt.title("BLUE")

plt.subplot(234), plt.imshow(H), plt.xticks([]), plt.yticks([]), plt.title("NUE")
plt.subplot(235), plt.imshow(S), plt.xticks([]), plt.yticks([]), plt.title("SATURATION")
plt.subplot(236), plt.imshow(V), plt.xticks([]), plt.yticks([]), plt.title("VALUE")

plt.show()
```

## 5. Скопіювати область інтересу (Region of Interest) чорно-білого зображення, використовуючи наступні діапазони 0:img.shape[0]/2, 0:img.shape[1]/2

```python
img_crop = img_gray[0:int(img_gray.shape[0]/2), 0:int(img_gray.shape[1]/2)]
plt.imshow(img_crop, cmap='gray'), plt.xticks([]), plt.yticks([]), plt.title("Region of Interest")
plt.show()
```

## 6. Зберегти область інтересу (Region of Interest).

```python
cv2.imwrite("OUT\RegionOfInterest.png", img_crop)
```
